```python
import pandas as pd
import numpy as np
```

```python
food = pd.read_csv("food.csv")
```

```python
print("\n--- Dataset Info ---")
print(food.info())
```

```python
print("\n--- Sample Rows ---")
display(food.head())
```

```python
print("\n--- Missing Values ---")
print(food.isnull().sum()[food.isnull().sum() > 0])
```

```python
food['Data.Household Weights.1st Household Weight Description'] = (
    food['Data.Household Weights.1st Household Weight Description'].fillna("Not Available")
)
food['Data.Household Weights.2nd Household Weight Description'] = (
    food['Data.Household Weights.2nd Household Weight Description'].fillna("Not Available")
)

print("\n‚úÖ Missing Values Filled")
```

```python
duplicate_rows = food.duplicated().sum()
print(f"\nüîÅ Duplicate Rows: {duplicate_rows}")
```

```python
food.columns = food.columns.str.replace(' ', '_').str.replace('.', '_').str.replace('-', '_')
print("\n‚úÖ Column names standardized")
```

```python
print("\n--- Cleaned Data Overview ---")
print(food.info())
print("\n--- Sample Data After Cleaning ---")
display(food.head())
```

```python
print("\n--- Descriptive Statistics ---")
display(food.describe(include='all'))
```

```python
food.to_csv("cleaned_food.csv", index=False)
print("\nüíæ Cleaned dataset saved as 'cleaned_food.csv'")
```
